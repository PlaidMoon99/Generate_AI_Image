```python
# Stable Diffusion 모델 로드 (Hugging Face 모델 허브에서 가져오기)
model_id = "runwayml/stable-diffusion-v1-5"
device = "cpu"  # CPU 사용

# 파이프라인 로드
pipe = StableDiffusionPipeline.from_pretrained(model_id)
pipe.to(device)
```

```python
# 텍스트 프롬프트를 이용하여 이미지 생성
prompt = "A fantasy landscape with mountains and rivers"
image = pipe(prompt).images[0]

```

## Login 'Hugging Face API' & Setting 'Stable Diffusion'

```python
# Login Hugging Face
from huggingface_hub import notebook_login
notebook_login()

# Import Library
import torch
from diffusers import StableDiffusionPipeline

# Stable Diffusion Model Load
model_id = 'runwayml/stable-diffusion-v1-5'
device = "cuda" if torch.cuda.is_available() else "cpu"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe.to(device)
```

## Writing Flask Server Code

```python
from flask import Flask, request, jsonify
from flask_ngrok import run_with_ngrok
import torch
from diffusers import StableDiffusionPipeline

app = Flask(__name__)
run_with_ngrok(app) # Connet to ngrok

# Loading Stable Diffusion Model (Using alreay roaded pipeline)
model_id = "runwayml/stable-diffusion-v1-5"
device = "cuda" if torch.cuda.is_available() else "cpu"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe.to(device)

@app.route('/generate_image', methods=['POST'])
def generate_image():
    # Get prompt from request
    data = request.get_json
    prompt = data.get('prompt') # Using prompt recieved by Client
    if prompt is none:
      return  jsonify({'error': 'Prompt is required'}), 400 # error return

    # Generate image using Stable Diffusion
    image = pipe(prompt).images[0]

    # Save image file
    image_path = "/images/generated_image.png" # Save to 'images' file & need to change name like adding  timestamp
    image.save(image_path)

    # Return image path to client
    return jsonify({'image_path': image_path})

if __name__ == '__main__':
    app.run()


try:
  pip = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
except Exception as e:
  print(f"Error loading model: {e}")
```
